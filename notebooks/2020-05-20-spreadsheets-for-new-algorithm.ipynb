```python
def parse_int_list(string):
    return tuple(map(int, string.split(',')))

def extract_dims(mat_name):
    main_parts = mat_name.split('-')
    m_part = main_parts[0]
    n_part = main_parts[-1]
    k_part = main_parts[-3].split('_')[-1]
    
    m_tuple, n_tuple, k_tuple = map(parse_int_list, [m_part, n_part, k_part])
    
    m, n, k = map(lambda v: np.product(v) ** 2, [m_tuple, n_tuple, k_tuple])
    
    return {"m": m, "n": n, "k": k,
            "m_dims": m_tuple, "n_tims": n_tuple, "k_dims": k_tuple}

def dims_df(names_column):
    return pd.DataFrame([extract_dims(name) for name in names_column],
                        index=names_column)
```

```python
dims = dims_df(times.index)[["m", "n", "k"]]
```

```python
raw_stats = utils.fetch("2020-05-20-matmul-dataset-stats")
```

```python
matrix_info = pd.concat(extraction.select_categories(raw_stats, ["mul"]).values())\
    [["key", "density", "transposed", "ones_a", "ones_b", "density_a", "density_b"]]\
    .rename({"key": "matrix",
            "density": "Density (C)",
            "transposed": "Transposed?",
            "ones_a": "Ones (A)",
            "ones_b": "Ones (B)",
            "density_a": "Density (A)",
            "density_b": "Density (B)"}, axis=1)\
    .set_index("matrix")
matrix_info.index = matrix_info.index.map(lambda s: s[len("matrices/"):])
matrix_info
```

```python
table = times.join([dims, matrix_info])
table.head()
```
